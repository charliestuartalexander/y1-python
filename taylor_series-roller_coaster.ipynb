## A. Taylor series approximations

One can approximate $\sin(x)$ using a finite number of terms of a Taylor series:

$$
\sin(x) \simeq \sum_{n=0}^N (-1)^n\frac{x^{2n+1}}{(2n+1)!} 
$$

One can also approximate the natural logarithm $\ln(1+x)$ using a finite number of terms of a Taylor series:

$$
\log(1+x) \simeq \sum_{n=1}^N (-1)^{n+1} \frac{x^n}{n}
$$

In this notebook we will explore these approximations, in particular how they depend on $N$. 





```python
# Functions to compute Taylor series approximations for sine and logarithm
# compute the approximation for all x simultaneously
def sinFunc(x,N):
    total = 0 * x
    for n in range(N+1):
        total += (-1)**n * x**(2*n+1)/np.math.factorial(2*n+1)
    return total
def logFunc(x,N):
    total = 0 * x
    for n in range(1,N+1):
        total += (-1)**(n+1) * (x**n)/n
    return total
```

The figure below shows sinx and approximations using Taylor series from N=0 to 6:

```python
# set the range of x

x1 = np.linspace(-3*np.pi,3*np.pi,101)

# call function for each N
sina=sinFunc(x1,0)
sinb=sinFunc(x1,1)
sinc=sinFunc(x1,2)
sind=sinFunc(x1,3)
sine=sinFunc(x1,4)
sinf=sinFunc(x1,5)
sing=sinFunc(x1,6)

# generate the exact sin value of x
y1 = np.sin(x1)

# plot the results
plt.ylim(-4,4)
plt.plot(x1,sina, '--', label="N=0")
plt.plot(x1,sinb, '--', label="N=1")
plt.plot(x1,sinc, '--', label="N=2")
plt.plot(x1,sind, '--', label="N=3")
plt.plot(x1,sine, '--', label="N=4")
plt.plot(x1,sinf, '--', label="N=5")
plt.plot(x1,sing, '--', label="N=6")
plt.plot(x1,y1, label='true',color='black')

plt.xlabel('x', fontsize=14)
plt.title('sin(x) and its approximation', fontsize=14)
plt.legend(fontsize=7)
plt.show()

```

The figure below shows logx+1 and approximations using Taylor series from N=1 to 7:

```python
# Generate plots comparing exact and Taylor approximations for the natural logarithm of x+1
# set the range of x

x2 = np.linspace(-0.9,2,101)

# call log function
loga=logFunc(x2,1)
logb=logFunc(x2,2)
logc=logFunc(x2,3)
logd=logFunc(x2,4)
loge=logFunc(x2,5)
logf=logFunc(x2,6)
logg=logFunc(x2,7)

# generate the exact log value of x
y2 = np.log(x2+1)

# plot the results
plt.ylim(-4,4)
plt.plot(x2,loga, '--', label="N=1")
plt.plot(x2,logb, '--', label="N=2")
plt.plot(x2,logc, '--', label="N=3")
plt.plot(x2,logd, '--', label="N=4")
plt.plot(x2,loge, '--', label="N=5")
plt.plot(x2,logf, '--', label="N=6")
plt.plot(x2,logg, '--', label="N=7")
plt.plot(x2,y2, label='true',color='black')


plt.xlabel('x', fontsize=14)
plt.title('log(x+1) and its approximation', fontsize=14)
plt.legend(fontsize=7)
plt.show()
```

### Error plots (Challenge section)

```python
# caclulate errors

errora=abs(y1-sina)
errorb=abs(y1-sinb)
errorc=abs(y1-sinc)
errord=abs(y1-sind)
errore=abs(y1-sine)
errorf=abs(y1-sinf)
errorg=abs(y1-sing)


# plot the results with small scale
plt.ylim(0,5)
plt.plot(x1,errora, '--', label="N=0")
plt.plot(x1,errorb, '--', label="N=1")
plt.plot(x1,errorc, '--', label="N=2")
plt.plot(x1,errord, '--', label="N=3")
plt.plot(x1,errore, '--', label="N=4")
plt.plot(x1,errorf, '--', label="N=5")
plt.plot(x1,errorg, '--', label="N=6")

plt.xlabel('x', fontsize=14)
plt.title('sin(x) and errors', fontsize=12)
plt.legend(fontsize=7)
plt.show()

# plot the results with a larger scale
plt.ylim(0,200)
plt.plot(x1,errora, '--', label="N=0")
plt.plot(x1,errorb, '--', label="N=1")
plt.plot(x1,errorc, '--', label="N=2")
plt.plot(x1,errord, '--', label="N=3")
plt.plot(x1,errore, '--', label="N=4")
plt.plot(x1,errorf, '--', label="N=5")
plt.plot(x1,errorg, '--', label="N=6")

plt.xlabel('x', fontsize=14)
plt.title('sin(x) and errors', fontsize=16)
plt.legend(fontsize=7)
plt.show()



```

As the number of terms (N) in the Taylor expansion increases,
the error decreases if x is close to 0.
However, overall if x is not close to 0, the error is larger for larger N values

```python
# set new range for x
x3=np.linspace(0.2,2*np.pi,101)
# note that log is used as a prefix to denote it being used for the log to log plot.
#call sin function with new range
logSina=sinFunc(x3,0)
logSinb=sinFunc(x3,1)
logSinc=sinFunc(x3,2)
logSind=sinFunc(x3,3)

#calculate new exact value and calculate error for each N
y3 = np.sin(x3)
logErrora=abs(y3-logSina)
logErrorb=abs(y3-logSinb)
logErrorc=abs(y3-logSinc)
logErrord=abs(y3-logSind)

#log to log plot
plt.loglog(x3,logErrora,'--', label="N=0")
plt.loglog(x3,logErrorb,'--', label="N=1")
plt.loglog(x3,logErrorc,'--', label="N=2")
plt.loglog(x3,logErrord,'--', label="N=3")

plt.xlabel('x', fontsize=14)
plt.title('log to log plot of sin(x) errors', fontsize=14)
plt.legend(fontsize=7)
plt.show()
```

As the lines seem to be straight as N increases, the log of the absolute error linearly depends on logx for larger N values.
Thus the absolute error can be modelled as $ 10^kx^m $ where k is the intercept and m is the gradient. From the graph as N increases, the lower the intercept and $10^k$ decreases thus making the error decrease for small x. From the graph, the greater N value chosen the greater the gradient thus if x is bigger $i.e: x>1$ then the overall absolute error increases for larger N.
